```python
from sklearn.decomposition import PCA  
from sklearn.datasets import load_iris  
import matplotlib.pyplot as plt  
```

```python
# 加载iris数据集  
iris = load_iris()  
X = iris.data  
y = iris.target  
```

```python
# 创建一个PCA对象，n_components指定了希望PCA降维后的特征维度数目  
pca = PCA(n_components=2)  
  
# 对数据进行PCA降维  
X_pca = pca.fit_transform(X)  
  
```

```python
# 输出降维后的数据形状  
print("原始数据形状:", X.shape)  
print("降维后数据形状:", X_pca.shape)  
  
# 绘制降维后的数据点  
plt.figure(figsize=(8, 6))  
colors = ['navy', 'turquoise', 'darkorange']  
for c, i, target_name in zip(colors, [0, 1, 2], iris.target_names):  
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], c=c, label=target_name, s=50)  
plt.xlabel('PCA feature 1')  
plt.ylabel('PCA feature 2')  
plt.legend(loc='best', shadow=False, scatterpoints=1)  
plt.title('PCA of IRIS dataset')  
plt.show()
# 查看组件（特征向量）  
print("PCA 组件:")  
print(pca.components_)  
  
# 解释组件：例如，查看第一个主成分中每个特征的权重  
print("第一个主成分中每个特征的权重:")  
print(pca.components_[0])  
  
# 查看变换后的数据（降维后的特征）  
print("变换后的数据（降维后的特征）:")  
print(X_pca)  
```
